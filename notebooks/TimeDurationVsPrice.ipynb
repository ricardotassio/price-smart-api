```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load the dataset as a CSV file
csv_file_path = 'transformed_ebay_data.csv'  # Replace with the actual file path
df = pd.read_csv(csv_file_path)

# Step 1: Visualize the relationship between Time Duration and Price
plt.figure(figsize=(10, 6))
plt.scatter(df['remainder__Time duration'], df['num__Price'], alpha=0.6, edgecolor='k')
plt.title('Scatter Plot: Time Duration vs. Price', fontsize=14)
plt.xlabel('Time Duration (remainder__Time duration)', fontsize=12)
plt.ylabel('Price (num__Price)', fontsize=12)
plt.grid(alpha=0.3)
plt.show()

# Step 2: Prepare data for regression analysis
X = df['remainder__Time duration']  # Predictor variable
y = df['num__Price']  # Response variable

# Add a constant to the predictor for the regression model intercept
X = sm.add_constant(X)

# Step 3: Fit the regression model
model = sm.OLS(y, X).fit()

# Step 4: Summarize and print the results
model_summary = model.summary()
print(model_summary)
```

Model Evaluation
Significance:

The relationship between Time Duration and Price is statistically significant but not practically significant, as indicated by the small coefficient and low R-squared.
Residuals:

The Durbin-Watson statistic (1.648) suggests no significant autocorrelation in residuals.
Condition Number: 778:

This indicates that multicollinearity is not a major concern in the model.

Conclusions
Impact of Time Duration:

While Time Duration has a statistically significant effect on price, the effect size is minimal, meaning it has very little practical impact.
Other Factors:

The low R-squared indicates that other factors (e.g., listing type, shipping type, seller reputation) are likely more influential in determining price.

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load the dataset as a CSV file
csv_file_path = 'transformed_ebay_data.csv'  # Replace with the actual file path
df = pd.read_csv(csv_file_path)

# Step 1: Visualize the relationship between Time Duration and Price
plt.figure(figsize=(10, 6))
plt.scatter(df['remainder__Time duration'], df['num__Price'], alpha=0.6, edgecolor='k')
plt.title('Scatter Plot: Time Duration vs. Price', fontsize=14)
plt.xlabel('Time Duration (remainder__Time duration)', fontsize=12)
plt.ylabel('Price (num__Price)', fontsize=12)
plt.grid(alpha=0.3)
plt.show()

# Step 2: Select additional predictors for the regression model
predictors = [
    'remainder__Time duration',
    'num__Feedback Score',
    'cat__Listing Type_Auction',
    'cat__Listing Type_AuctionWithBIN',
    'cat__Listing Type_FixedPrice',
    'cat__Listing Type_StoreInventory',
    'cat__Shipping Type_Calculated',
    'cat__Shipping Type_CalculatedDomesticFlatInternational',
    'cat__Shipping Type_Flat'
]

# Step 3: Prepare the data for multiple regression analysis
X_multi = df[predictors]
y_multi = df['num__Price']

# Add a constant to the predictors for the intercept
X_multi = sm.add_constant(X_multi)

# Step 4: Fit the multiple regression model
multi_model = sm.OLS(y_multi, X_multi).fit()

# Step 5: Summarize and print the regression results
multi_model_summary = multi_model.summary()
print(multi_model_summary)
```
