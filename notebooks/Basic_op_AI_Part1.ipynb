```python
# Convert non-missing values to 1 and missing values to 0 in a â€™Gallery URLâ€˜ column
# Convert non-missing values to 1 and missing values to 0 in a â€™Large Image URLâ€˜ column
# Convert non-missing values to 1 and missing values to 0 in a â€™Super Size Image URLâ€˜ column
df['Store URL_flag'] = df['Store URL'].notna().astype(int)
df['Large Image URL_flag'] = df['Large Image URL'].notna().astype(int)
df['Super Size Image URL_flag'] = df['Super Size Image URL'].notna().astype(int)
```

```python
#One-Hot Encoding for Listing Type (manually)
listing_type_categories = df['Listing Type'].unique()
for category in listing_type_categories:
    df[f'Listing Type_{category}'] = (df['Listing Type'] == category).astype(int)

#One-Hot Encoding for Shipping Type (manually)
shipping_type_categories = df['Shipping Type'].unique()
for category in shipping_type_categories:
    df[f'Shipping Type_{category}'] = (df['Shipping Type'] == category).astype(int)

#Label Encoding for Top Rated Seller (binary column: True/False)
label_encoder = LabelEncoder()
df['Top Rated Seller'] = label_encoder.fit_transform(df['Top Rated Seller'])

#One-Hot Encoding for Condition (manually)
condition_categories = df['Condition'].unique()
for category in condition_categories:
    df[f'Condition_{category}'] = (df['Condition'] == category).astype(int)

#Ensure that the currency column is set to USD for all rows
df['Currency'] = 'USD'

#Drop original categorical columns as we have encoded them
df.drop(columns=['Listing Type', 'Shipping Type', 'Condition'], inplace=True)

#Show the transformed DataFrame
print(df)
```

```python
# List of numerical columns to scale
numerical_columns = ['Price', 'Feedback Score']

# Define the ColumnTransformer
column_transformer = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_columns)  # Scaling for numerical columns
    ],
    remainder='passthrough'  # Keep untransformed columns unchanged
)

# Apply transformation
final = column_transformer.fit_transform(df)

# Get column names, including original ones for 'passthrough' columns
column_names = column_transformer.get_feature_names_out()

# Convert the result back to a DataFrame with proper column names
df_cleaned = pd.DataFrame(final, columns=column_names)

# Show the transformed DataFrame
print(df_cleaned.head())
```
