```python
import os
```

```python
def load_csv_from_file(folder_path):
    combined_df = pd.DataFrame()
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(folder_path, filename)
            if os.path.getsize(file_path) > 0: 
                try:
                    df = pd.read_csv(file_path)
                    combined_df = pd.concat([combined_df, df], ignore_index=True)
                except pd.errors.EmptyDataError:
                    print(f"Skipping empty or malformed file: {filename}")
            else:
                print(f"Skipping empty file: {filename}")
    return combined_df
```

```python
# encoding the catergorical varibles
```

```python
catogory_columns = ['Listing Type','Shipping Type','Top Rated Seller','Condition'] 
```

```python
for category in catogory_columns:
    print(f'{category}  :  \n {combined_df[category].unique()}')
```

```python
# encoding choices
# 	•	Listing Type: Use One-Hot Encoding since this column has a few categories with no ordinal relationship.
# 	•	Shipping Type: Use One-Hot Encoding since there are multiple categories, and they don’t have a specific order.
# 	•	Top Rated Seller: Use Label Encoding since this column is binary (True/False).
# 	•	Condition: Use One-Hot Encoding as there are many categories, and they represent different conditions without any specific order.
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer


# Define the ColumnTransformer
column_transformer = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Price', 'Feedback Score']),  # Scaling for numerical columns
        ('cat', OneHotEncoder(), catogory_columns)               # One-Hot Encoding for categorical columns
    ],
    remainder='passthrough'  # Keep untransformed columns unchanged
)

# Apply transformation
final = column_transformer.fit_transform(combined_df)

# Get column names, including original ones for 'passthrough' columns
column_names = column_transformer.get_feature_names_out()

# Convert the result back to a DataFrame with proper column names
df_final = pd.DataFrame(final, columns=column_names)
```
